Цель работы: реализовать базовый анализ датасета при помощи библиотеки panpas (Вывести весь датасет и описать его, вывести размерноть массива, вывести наименование всех колонок, вывести все уникальные значения, отсортировать по определенным параметрам, удалить ненужные столбцы или строчки, заменить определенные значения в датасете, удалить дубликаты, провести анализ с помощью функций info и describe, провести выборку данных по строкам и столбцам с помощью loc, сохранить новый датасет)

```python
import pandas as pd

anime = pd.read_csv("anime.csv")

anime
```

Для начала нам необходимо прочитать csv фаил, записать его в переменную и вывести. После вывода мы можем увидеть все элементы, содержащиеся в нашем датасете

```python
anime.shape
```

Далее необходимо узнать размерность нашего датасета с помощью функции shape. В нашем случае у нас 12294 строчки и 7 столбцов. 

```python
anime.columns
```

После с помощью функции columns выведим названия всех столбцов. У нас столбы anime_id - просто какой-то айдишник, name - название аниме, genre - жанр аниме, type - тип аниме (что-то вроде фильм, сериал и т.п.), episodes - количество эпизодов, rating - рейтинг аниме на какой-то площадке, members - количество просмотров.

```python
for col in anime:
    print(anime[col].unique())
```

Далее при помощи функции unique мы выведим все уникальные значения, содержащиеся в нашем датасете. 

```python
anime[anime["type"] == "TV"]
```

При помощи данной команды мы можем отсортировать элементы датасета по определенному параметру и посмотреть все необходимые нам элементы.

```python
anime = anime[anime["genre"] != "Hentai"]

anime
```

В данном случае мы перезапишем все элементы датасета кроме тех элементов с типом хентай. Тем самым мы удалим ненужные строчки. 

```python
import numpy as np
anime["rating"] = np.where(anime["rating"].isnull(), "not info", anime["rating"])

anime
```

Здесь мы заменим значения NaN в с столбце рейтинга на not info

```python
anime.drop_duplicates()

anime
```

Тут мы удаляем все повторяющиеся элементы (которых у нас нет, но пункт должен быть :))

```python
anime.info()
```

После мы выведим подробную информацию о всех столбцах, в которой мы можем наблюдать, что столбцы anime_id и members имеют тип данных int, а все остальные object. Также не все столбцы заполнены значениями, есть те, у которых количество заполненых ячеек меньше чем количество столбцов.

```python
anime.describe(include='all')
```

Далее с помощью функции describe выведим подробную статиску по нашему датасету (количество непустых ячеек, количество уникальных значений столбцов, самые часто встречающиеся значения, среднее арифметическое значение, стандартное отклонение и т.д.)

```python
data = anime.loc[[1, 2, 11000, 5000, 7777, 5555], ["name", "type", "episodes", "rating"]]

data
```

Здесь мы выводим строки с конкретным номером (1, 2, 11000, 5000, 7777, 5555) и конкретные столбцы у этих строк (name, type, episodes, rating). И мы можем увидеть, что аниме под номером 1 - это Стальной алхимик: Братство, с типом TV, 64 эпизодами и рейтингом 9.26 (очень класная анимешка, советую посмотреть).

```python
anime.to_csv("anime_new.csv", index=False)
```
